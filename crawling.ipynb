## 크롤링한 내용 list에 넣어주기

```python
URL = 'https://www.ctis.re.kr/ko/techClass/classification.do?key=1141'
driver = webdriver.Chrome(executable_path='chromedriver')
driver.get(url=URL)

for i in range(46):
    s = '#table_box > table > tbody > tr:nth-child('+str(i)+') > td.bgw > a'

    for label in driver.find_elements_by_css_selector(s):
        print(label.text)
        label.send_keys(Keys.ENTER)
        time.sleep(1) # 너무 빨리 켜져서 크롤링 못할때가 있어서 1초 쉬어주기
        
        if(i>40):
            korean = driver.find_elements_by_css_selector('#pdfShows > dl > dd:nth-child(4) > table > tbody > tr:nth-child(1) > td')[0].text
            english = driver.find_elements_by_css_selector('#pdfShows > dl > dd:nth-child(4) > table > tbody > tr:nth-child(2) > td')[0].text
        else :    
            korean = driver.find_elements_by_css_selector('#pdfShows > dl > dd:nth-child(4) > table > tbody > tr:nth-child(1) > td')[0].text
            english = driver.find_elements_by_css_selector('#pdfShows > dl > dd:nth-child(4) > table > tbody > tr:nth-child(2) > td')[0].text
        
        # white space 지워주기
        korean = korean.replace(" ","")
        english = english.replace(" ", "")
        
        koreanSplit = korean.split(',')
        englishSplit = english.split(',')
        
        for tag in koreanSplit:
            if tag not in koreanLists[i]:
                koreanLists[i].append(tag)
                
        for tag in englishSplit:
            if tag not in englishLists[i]:
                englishLists[i].append(tag)        
        
        driver.get(url=URL)
```

```python
train_ko = []
train_en = []
```

```python
import re
```

```python
for i, j, k in zip(train_set['요약문_한글키워드'], train_set['요약문_영문키워드'], train_set['label']):

    ko = i.replace("\n", "")
    en = j.replace("\n", "")
    
    ko = ko.replace(" ", "")
    en = en.replace(" ", "")
    
    koSplit = re.split("[\|,]", ko)
    enSplit = re.split("[\|,]", en)
    
    for m in koSplit:
        if(m!=''):
            if(train_ko[k].get(m)==None):
                train_ko[k][m]=1
            else:
                train_ko[k][m]=train_ko[k][m]+1
            
    for m in enSplit:
        if(m!=''):
            if(train_en[k].get(m)==None):
                train_en[k][m]=1
            else:
                train_en[k][m]=train_en[k][m]+1    
    
```

# 비율 확인하기

```python
# train_ko, train_en // list안에 dict
# 소분류 koreanLists, englishLists // list안에 list
# 중분류 middleKorean, middleEnglish // list안에 list
```

```python
for i in range(46):
    count_keyword = 0
    count_train = 0
    
    temp = train_ko[i].values()
    for k in temp:
        count_train = count_train+k
        
    for j in koreanLists[i]:
        if(train_ko[i].get(j)!=None):
            count_keyword += train_ko[i].get(j)
            
    print(i,"번째",count_keyword/count_train*100,"%")
```
