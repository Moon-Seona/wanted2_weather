## 크롤링한 내용 list에 넣어주기

```python
URL = 'https://www.ctis.re.kr/ko/techClass/classification.do?key=1141'
driver = webdriver.Chrome(executable_path='chromedriver')
driver.get(url=URL)

for i in range(46):
    s = '#table_box > table > tbody > tr:nth-child('+str(i)+') > td.bgw > a'

    for label in driver.find_elements_by_css_selector(s):
        print(label.text)
        label.send_keys(Keys.ENTER)
        time.sleep(1) # 너무 빨리 켜져서 크롤링 못할때가 있어서 1초 쉬어주기
        
        if(i>40):
            korean = driver.find_elements_by_css_selector('#pdfShows > dl > dd:nth-child(4) > table > tbody > tr:nth-child(1) > td')[0].text
            english = driver.find_elements_by_css_selector('#pdfShows > dl > dd:nth-child(4) > table > tbody > tr:nth-child(2) > td')[0].text
        else :    
            korean = driver.find_elements_by_css_selector('#pdfShows > dl > dd:nth-child(4) > table > tbody > tr:nth-child(1) > td')[0].text
            english = driver.find_elements_by_css_selector('#pdfShows > dl > dd:nth-child(4) > table > tbody > tr:nth-child(2) > td')[0].text
        
        # white space 지워주기
        korean = korean.replace(" ","")
        english = english.replace(" ", "")
        
        koreanSplit = korean.split(',')
        englishSplit = english.split(',')
        
        for tag in koreanSplit:
            if tag not in koreanLists[i]:
                koreanLists[i].append(tag)
                
        for tag in englishSplit:
            if tag not in englishLists[i]:
                englishLists[i].append(tag)        
        
        driver.get(url=URL)
```
