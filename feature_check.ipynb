```python
# train['요약문_한글키워드'][0].split('[ ,|]')
re.split(' ,|,| \| ', train['요약문_한글키워드'][36])
```

```python
feature = '요약문_한글키워드'
keymap = {}
nanlist = []
for i in tqdm(range(len(train[feature]))): # 시작
    key = train.loc[i]['label']
    if(key not in keymap.keys()): # 키가 없으면 만들어라
        keymap[key] = {}
    try: # 키워드가 있으면 자르고
        st = train[feature][i].replace(" ", "")
        st = st.replace("|", ",")
        li = st.split(',')
    except: # 없으면 nanlist에 넣어라
        nanlist.append(i)
        continue
    for j in li:
        if(j == ''):
            continue
        try:
            keymap[key][j] = keymap[key][j] + 1
        except:
            keymap[key][j] = 1;
```

```python
# 매 label 기준 3% 이상인 친구들은 새로운 dict에 넣어준다.
key_key = {}
for i in range(46):
    key_key[i] = []
    s = sum(keymap[i].values())
    temp = sorted(keymap[i].items(), key = lambda item: item[1], reverse=True) # list
    for j in range(len(temp)):
        if(temp.values()[:j] / s >= 0.03):
            key_key[i].append(temp.keys()[:j])
        else:
            break
    if(len(key_key[i]) == 0):
        key_key[i].append(temp[i][0])
```

## 3% keyword 빼내기

```python
for i in range(0, 46):
    keymap[i] = sorted(keymap[i].items(), key=lambda x:x[1], reverse=True)

# {[()]} 형식으로 되어있음.
```

```python
keyword3Per = [0 for i in range(0, 46)]

for i in range(1, 46):
    temp = keymap[i]
    keyword3Per[i] = []
    
    value_sum = 0
    value_count = 0
    key_count = 0
    
    for j in temp:
        value_sum += j[1]
    
    for j in temp:
        value_count+=j[1]
        print(i, " ", value_count/value_sum)
        
        if(value_count/value_sum > 0.03):
            if key_count==0:
                keyword3Per[i].append(j)
                break;
            else:
                break;
            
        if(value_count/value_sum < 0.03):
            keyword3Per[i].append(j)
        
        
            
        key_count+=1
        
```

```python
for i in range(1, 46):
    print(keyword3Per[i])
```
